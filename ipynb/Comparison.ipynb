```python
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pylab as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns

import math
import sys
%matplotlib inline
sns.set()

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
def rmse(test,pred):
    mse = mean_squared_error(test, pred)
    rmse = np.sqrt(mse)
    return rmse

# r21 = r2_score(test, pred)
```

## training : 730days ,  test : 60days

```python
rmse_list = []
r2_list = []

for i in range(1,6):
    pred_pass = '../result/2test' + str(i) + '/result_mean.csv' 
    pred = pd.read_csv(pred_pass)
    pred = pred.iloc[:,1:]
    true_pass = '../test/2test' + str(i) + '.csv' 
    true = pd.read_csv(true_pass)
    true = true["CustNum"]
    
    rmse_value = rmse(true,pred)
    rmse_list.append(rmse_value)
    
    r2_value = r2_score(true, pred)
    r2_list.append(r2_value)
```

```python
date_list = []
for i in range(1,6):
    date_pass = '../test/2test' + str(i) + '.csv' 
    date = pd.read_csv(date_pass)['date']
    date_list.append(date[0])
```

```python
import datetime
from matplotlib import pylab as plt


fig, ax = plt.subplots(figsize=(20, 10))
plt.plot(date_list, rmse_list, label='BSTS')
# plt.plot(dayadd_pred,label='Predicted')
# plt.fill_between(test_index,Interval2.iloc[:,0], Interval2.iloc[:,1], color='b', alpha=0.2,label='95% error')

plt.xlabel("Date",fontsize=25)
plt.ylabel("RMSE",fontsize=25)
plt.tick_params(length=0)
plt.legend(fontsize=20,ncol=2)
plt.tick_params(labelsize=30)
plt.xticks(rotation=45,fontsize=17)
# fig.savefig("lgbm_predict.png",bbox_inches="tight")

# plt.plot(Interval2.iloc[:,1], 'k:')
# plt.plot(Interval2.iloc[:,0], 'k:')
```

```python
def make_rmse(used_variable):
    rmse_list2 = []
    r2_list2 = []
    for i in range(1,23):
        # pred_pass2 = '../result/skt/test' + str(i) + '/result_mean.csv' 
        pred_pass2 = '../result/' + used_variable + '/test' + str(i) + '/result_mean.csv' 
        pred2 = pd.read_csv(pred_pass2)
        pred2 = pred2.iloc[:,1:]
        true_pass2 = '../test/test' + str(i) + '.csv' 
        true2 = pd.read_csv(true_pass2)
        true2 = true2["CustNum"]
        
        rmse_value2 = rmse(true2,pred2)
        rmse_list2.append(rmse_value2)
        
        r2_value2 = r2_score(true2, pred2)
        r2_list2.append(r2_value2)
    return rmse_list2, r2_list2
```

```python
none_rmse, none_r2 = make_rmse("non")
trend_rmse, trend_r2 = make_rmse("trend")
ei_rmse, ei_r2 = make_rmse("ei")
ei_trend_rmse, ei_trend_r2 = make_rmse("ei_trend")
```

```python
data_list = [none_rmse, ei_rmse, none_r2, ei_r2]
```
